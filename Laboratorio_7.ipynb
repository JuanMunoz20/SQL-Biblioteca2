```python
import numpy as np

# Crear un array de n炭meros del 1 al 10
array = np.arange(1, 11)

# Calcular la suma, la media y el valor m叩ximo
suma = np.sum(array)
media = np.mean(array)
maximo = np.max(array)

# Imprimir los resultados
print("Suma:", suma)
print("Media:", media)
print("Valor m叩ximo:", maximo)
```

```python
import numpy as np

# Crear un array de 3x4 lleno de ceros
array_zeros = np.zeros((3, 4))

# Cambiar la forma del array a 2x6
array_reshape = array_zeros.reshape(2, 6)

# Imprimir el array con la nueva forma
print(array_reshape)
```

```python
import numpy as np

# Generar una matriz aleatoria de 3x3
matriz_aleatoria = np.random.rand(3, 3)

# Calcular la transpuesta
matriz_transpuesta = matriz_aleatoria.T

# Imprimir la matriz original y la transpuesta
print("Matriz original:")
print(matriz_aleatoria)
print("\nMatriz transpuesta:")
print(matriz_transpuesta)
```

```python
import numpy as np

# Crear dos arrays con 5 valores cada uno
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([6, 7, 8, 9, 10])

# Concatenar los arrays
array_concatenado = np.concatenate((array1, array2))

# Imprimir el array concatenado
print(array_concatenado)
```
